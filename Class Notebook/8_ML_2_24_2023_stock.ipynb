# Data

```python
START = '2016-01-01'
END   = '2019-12-31'
```

```python
import yfinance as yf
yf.Ticker('AMZN').history(start=START, end=END).head()
```

```python
import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
data
```

```python
type(data)
```

```python
len(data)
```

```python
data[1]
```

```python
stock_list = data[0]['Symbol'].values
stock_list
```

```python
part_stock_list = stock_list[:100]
```

```python
df = pd.DataFrame()
for t in part_stock_list:
  df[t] = yf.Ticker(t).history(start=START, end=END)['Close']
```

```python
df.head().round()
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.index = [i.date() for i in df.index]
df.head().round()
```

# Correlations

```python
df_corr = df.corr()
```

```python
df_corr
```

```python
import seaborn as sns
sns.heatmap(df_corr);
```

```python
for row in df_corr.index:
  for col in df_corr.columns:
    if row != col:
      if df_corr.loc[row, col] > 0.98:
        print(f'{row} -- {col} -- {df_corr.loc[row, col]:.2f}')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(df['ADBE'], label='ADBE')
plt.plot(df['AMZN'], label='AMZN')
plt.legend();
```

```python
plt.figure(figsize=(20,5))
plt.plot(df['ADBE']/max(df['ADBE']), label='ADBE')
plt.plot(df['AMZN']/max(df['AMZN']), label='AMZN')
plt.legend();
```

```python
import pandas as pd
df = pd.DataFrame(dataset.data)
df['Target'] =dataset.target
df.head()
```

```python
plt.figure(figsize=(20,8))
for i in range(len(t_names)):
  plt.subplot(5, 7, i+1)
  index = df[df['Target']==i].index[0]
  plt.title(t_names[dataset.target[index]])
  plt.imshow(dataset.images[index], 'gray')
  plt.axis('off');
plt.tight_layout()
```

```python
# first 30 pictures
df[df['Target']==8][:30]
```

```python
df_elim = pd.DataFrame()
for i in range(len(t_names)):
  df_elim = pd.concat( [ df_elim, df[df['Target']==i][:30] ] )
```

```python
df_elim.head()
```

```python
df_elim.Target.value_counts()
```

```python
X = df_elim.iloc[:,:-1].values
y = df_elim['Target'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train.shape
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(X_train)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_));
```

```python
np.cumsum(pca.explained_variance_ratio_)[100]
```

```python
pca100 = PCA(n_components=100)
X_train_pca = pca100.fit_transform(X_train)
X_test_pca = pca100.transform(X_test)
```

```python
X_train_pca.shape
```

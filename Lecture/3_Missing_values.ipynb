# Imputation Methods

## Filling with a constant number

```python
df_stock.fillna(999)
```

## Filling with a string

```python
df_stock.fillna('EMPTY')
```

## Filling with forward fill

```python
df_stock.fillna( method ='ffill')
```

## Filling with backward fill

```python
df_stock.fillna( method ='bfill')
```

## Filling with mean

```python
df_stock.fillna(df_stock.mean())
```

## Filling with median

```python
df_stock.fillna(df_stock.median())
```

# Imputer Methods

## Simple Imputer

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')   # mean, median, most_frequent, constant
```

```python
imputer.fit(df_stock)
imputer.transform(df_stock)
```

```python
pd.DataFrame(imputer.transform(df_stock), columns=df_stock.columns, index=df_stock.index)
```

## knn Imputer

imputed using the mean value from n_neighbors nearest neighbors found in the training set
- https://scikit-learn.org/stable/modules/generated/sklearn.impute.KNNImputer.html


```python
from sklearn.impute import KNNImputer
imputer_knn = KNNImputer(n_neighbors=2)
imputer_knn.fit_transform(df_stock)
```

```python
pd.DataFrame(imputer_knn.fit_transform(df_stock), columns=df_stock.columns, index=df_stock.index)
```

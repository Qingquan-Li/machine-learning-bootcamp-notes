# Unsupervised Learning

- No output (label) just the input
- Create a new representation of the data
  - easier to understand compared to the original representation of the data


- AIM
  - finding underlying (hidden) structures or distributions in the data

- TASKS

  - clustering
    - partition data into distinct groups of similar items.

  - dimension reduction
    - summarizes the essential characteristics of the data with fewer features.

  - outlier detection


- EXAMPLES
  - Topic extraction from a collection of text documents (like reviews)
  - Organization of pictures

- DISADVANTAGES
  - No way to measure the success (right output is unknown)

- APPLICATIONS
  - often used in an exploratory setting to understand the data better
  - preprocessing step for supervised algorithms

# Scaling

- ML algorithms which use distance between data points require scaled data


## Minmax Scaler
- scales values between 0 (min) and 1 (max)

- FORMULA
   - $scaled(data) =\frac{data-min}{max-min}$

- Scale the  Training and Test Data in the Same Way

```python
# data
data = [[1],[2],[3],[4],[5]]
```

```python
# import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data)
```

```python
# transform data
scaler.transform(data)
```

```python
# transform 2.5
scaler.transform([[2.5]])
```

# Kmeans
- clusters data by trying to separate samples into given number of groups
- each sample belongs to the cluster with the nearest mean (centroid)
STEPS
- choose the intial centroids (randomly or given)
- Apply the following two steps iteatively

  1. assign each sample to its nearest centroid
  2. set new centroids as the mean of each cluster
- MINIMIZE  
$inertia = \sum_{i=1}^k\sum_{x\in C_i} |x-m_i|^2$  for the clusters $C_1,C_2,...,C_k$


- ELBOW RULE  
the point in the curve where the inertia stops dropping fast (the “elbow”)

## kMeans and Iris data

```python
# scale the iris data
# apply min_max_scaler to training and test set
X_iris_train_scaled = scaler.fit_transform(X_iris_train)
X_iris_test_scaled  = scaler.transform(X_iris_test)
```

```python
# Iris in Supervised and Unsupervised Learning
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.title('Scaled Iris in Supervised Learning')
train_targets_iris = [t_names_iris[i] for i in y_iris_train]
sns.scatterplot( x=X_iris_train_scaled[:,0], y=X_iris_train_scaled[:,1], hue=train_targets_iris, palette='bright' )
plt.xlabel(f_names_iris[0])
plt.ylabel(f_names_iris[1])
plt.subplot(1,2,2)
# Iris in Unsupervised Learning
plt.title('Scaled Iris in Unsupervised Learning')
sns.scatterplot( x=X_iris_train_scaled[:,0], y=X_iris_train_scaled[:,1], color='black')
plt.xlabel(f_names_iris[0])
plt.ylabel(f_names_iris[1]);
```

```python
from sklearn.cluster import KMeans
# inertia graph
l_inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_iris_train[:,:2])
    l_inertia.append(kmeans.inertia_)
plt.plot( range(1,11), l_inertia )
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid();
```

```python
# build the clustering model with 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_iris_train_scaled[:,:2])
```

```python
kmeans.labels_[:5]
```

```python
kmeans.cluster_centers_.shape
```

```python
# Iris in Supervised and Unsupervised Learning
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
plt.title('Scaled Iris in Supervised Learning')
sns.scatterplot( x=X_iris_train_scaled[:,0], y=X_iris_train_scaled[:,1], hue=train_targets_iris, palette='bright' )
plt.xlabel(f_names_iris[0])
plt.ylabel(f_names_iris[1])
plt.subplot(2,2,2)
# Iris in Unsupervised Learning
plt.title('Scaled Iris in Unsupervised Learning')
sns.scatterplot( x=X_iris_train_scaled[:,0], y=X_iris_train_scaled[:,1], color='black')
plt.xlabel(f_names_iris[0])
plt.ylabel(f_names_iris[1])
plt.subplot(2,2,3)
# Iris in kMeans Learning
plt.title('Scaled Iris in Supervised Learning')
sns.scatterplot( x=X_iris_train_scaled[:,0], y=X_iris_train_scaled[:,1], hue=train_targets_iris, palette='bright' )
sns.scatterplot( x = kmeans.cluster_centers_[:,0], y  = kmeans.cluster_centers_[:,1], color = 'r', marker = '*', s=600 );
plt.xlabel(f_names_iris[0])
plt.ylabel(f_names_iris[1])
plt.subplot(2,2,4)
# Iris in kMeans Learning
plt.title('Scaled Iris in kMeans Learning')
sns.scatterplot( x=X_iris_train_scaled[:,0], y=X_iris_train_scaled[:,1], hue=kmeans.labels_ ,palette='bright')
sns.scatterplot( x = kmeans.cluster_centers_[:,0], y  = kmeans.cluster_centers_[:,1], color = 'r', marker = '*', s=600 );
plt.xlabel(f_names_iris[0])
plt.ylabel(f_names_iris[1])
plt.tight_layout;
```

# Principal Component Analysis
- commonly used for dimensionality reduction and visualization
- change of base


```python
import numpy as np
np.random.seed(42)
plt.figure(figsize=(5,5))
x = np.linspace(-1,1,20)
noise = np.random.randn(20)/3
y = 2*x+noise
y = y-np.mean(y)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(np.stack([x,y]).T)
m1 = pca.components_[0,1]/pca.components_[0,0]
m2 = pca.components_[1,1]/pca.components_[1,0]
y1,y2 = m1*x, m2*x
plt.scatter(x,y,c='g');
plt.plot(x,y1, 'r--', label='PC1')
plt.plot(x,y2, 'b--', label='PC2')
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.grid()
plt.legend();
```

```python
# use plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots()

fig.add_trace(go.Scatter(x=x, y=y1, line=dict(color='red',  dash='dash'),  name="PC1")  )
fig.add_trace(go.Scatter(x=x, y=y2, line=dict(color='blue',  dash='dash'),  name="PC2")  )
fig.add_trace(go.Scatter(x=x, y=y,  mode='markers',marker=dict(size=20, color='green'),  name="data")  )

fig.update_xaxes(range=(-2,2))
fig.update_yaxes(range=(-2,2))

fig.update_layout(title_text="PCA",autosize=False, width=600, height=600)
```

```python
# use plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots()

fig.add_trace(go.Scatter(x=[-2,2], y=[0,0], line=dict(color='red',  dash='dash'),  name="PC1")  )
fig.add_trace(go.Scatter(x=[0,0], y=[-2,2], line=dict(color='blue',  dash='dash'),  name="PC2")  )
fig.add_trace(go.Scatter(x=pca.transform(np.stack([x,y]).T)[:,0], y=pca.transform(np.stack([x,y]).T)[:,1],marker=dict(size=15, color='green'),  mode='markers',  name="tranformed data")  )

fig.update_xaxes(range=(-2,2))
fig.update_yaxes(range=(-2,2))

fig.update_layout(title_text="Transformed Data",autosize=False, width=600, height=600)
```

```python
# use plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots()

fig.add_trace(go.Scatter(x=[-2,2], y=[0,0], line=dict(color='red',  dash='dash'),  name="PC1")  )
fig.add_trace(go.Scatter(x=[0,0], y=[-2,2], line=dict(color='blue',  dash='dash'),  name="PC2")  )
fig.add_trace(go.Scatter(x=pca.transform(np.stack([x,y]).T)[:,0], y=[0]*20,  mode='markers', marker=dict(size=20, color='green'), name="reduced data")  )

fig.update_xaxes(range=(-2,2))
fig.update_yaxes(range=(-2,2))

fig.update_layout(title_text="Data After Dimension Reduction",autosize=False, width=600, height=600)
```
